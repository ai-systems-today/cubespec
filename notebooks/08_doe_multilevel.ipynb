## 1. When 2 levels is not enough

For curvature, you need at least 3 levels per factor. The dashboard accepts up to 5. Mixed-level designs are full-factorial only — fractional designs are locked because aliasing rules differ for ≠ 2 level factors.

We pick a 3-factor sub-design with 3, 3, 5 levels (45 runs) and fit a quadratic effect on P9 to expose the curvature in P0_rho.


```python
from cubespec import DEFAULT_CSP, full_factorial
import numpy as np, pandas as pd

# 3 evenly-spaced levels per factor across all 7 inputs (3**7 = 2187 runs).
# This mirrors the dashboard's 'multi-level' DOE mode (full factorial only).
df = full_factorial(DEFAULT_CSP, levels=3, span_sds=1.0)
print(f'rows = {len(df)}  (3^7 = 2187)')
df.head()
```

## 2. Curvature signature in P0_rho


```python
import matplotlib.pyplot as plt
g = df.groupby('P0_rho')['P9_compressive_strength'].mean().reset_index()
fig, ax = plt.subplots(figsize=(6.5, 3.6))
ax.plot(g.P0_rho, g.P9_compressive_strength, 'o-', color='#5b8def')
ax.set_xlabel('P0_rho (kg/m³)')
ax.set_ylabel('mean P9 across other factors (MPa)')
ax.set_title('3 levels reveal curvature that 2 levels would miss')
fig.tight_layout(); plt.show()
```

## 3. Why fractional gets locked

The dashboard greys out fractional designs as soon as any factor has > 2 levels. Reason: the 2^(k-p) generator algebra requires every factor to be a ±1 contrast. A 3-level factor has no single ± representation — so the design must be full.
