## 1. Full 2⁷ factorial

Setting every factor to ±1σ around its mean and enumerating all combinations gives `2⁷ = 128` runs. We compute the main effects on all three outputs and rank them.


```python
from cubespec import DEFAULT_CSP, full_factorial, main_effects
import pandas as pd
df_full = full_factorial(DEFAULT_CSP, levels=2)
print(f'design has {len(df_full)} runs across {df_full.filter(like="_coded").shape[1]} factors')
df_full.head()
```

```python
eff = main_effects(df_full)
p9 = eff.query("output == 'P9_compressive_strength'").sort_values('abs', ascending=False)
p9
```

## 2. Tornado chart for P9


```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(6.5, 3.6))
ax.barh(p9.factor[::-1], p9.effect[::-1], color='#5b8def')
ax.set_xlabel('main effect on P9 (MPa)')
ax.set_title('Full 2⁷ — ranked main effects on P9')
fig.tight_layout(); plt.show()
```

## 3. Cross-output ranking

The same factor can be the dominant driver for one output and irrelevant for another. Tabulate the top-3 per output.


```python
for out in ('P7_def', 'P8_strain', 'P9_compressive_strength'):
    sub = eff.query('output == @out').sort_values('abs', ascending=False).head(3)
    print(f'\nTop drivers for {out}')
    print(sub.to_string(index=False))
```
