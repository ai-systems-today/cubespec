# 05 — Bootstrap CI on the P9 mean


```python
from cubespec import DEFAULT_CSP, sample_independent, compute_outputs_batch, bootstrap_mean_ci
Y = compute_outputs_batch(sample_independent(DEFAULT_CSP, n=20_000, seed=7))[:, 2]
for B in (200, 500, 1000, 2000):
    r = bootstrap_mean_ci(Y, B=B, seed=1337)
    print(f'B={B:5d}  CI=[{r.lo:.4f}, {r.hi:.4f}]  width={r.hi - r.lo:.4f}')
```
