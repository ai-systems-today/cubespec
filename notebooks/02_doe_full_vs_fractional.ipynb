# 02 — Full 2⁷ vs fractional 2^(7-3)


```python
from cubespec import DEFAULT_CSP, full_factorial, fractional_factorial, main_effects
full = full_factorial(DEFAULT_CSP, levels=2)
frac = fractional_factorial(DEFAULT_CSP, fraction='1/8')
print('Full runs:', len(full), '| Fractional 1/8 runs:', len(frac))
print(main_effects(full).query("output == 'P9_compressive_strength'").sort_values('abs', ascending=False))
```
