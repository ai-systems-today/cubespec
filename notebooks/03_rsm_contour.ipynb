# 03 — RSM contour for (P4_Fx, P1_dx)


```python
import numpy as np, matplotlib.pyplot as plt
from cubespec import DEFAULT_CSP, sample_independent, compute_outputs_batch
from cubespec.rsm import fit_quadratic, predict_grid
X = sample_independent(DEFAULT_CSP, n=2000, seed=42)
Y = compute_outputs_batch(X)[:, 2]
model = fit_quadratic(X, Y)
print('R² =', model.r2)
base = X.mean(axis=0)
A, B, Z = predict_grid(model, base, factor_a=4, factor_b=1, span_a=(9.5e5, 1.05e6), span_b=(149, 151))
plt.contourf(A, B, Z, levels=20); plt.colorbar(label='P9 (MPa)')
plt.xlabel('P4_Fx (N)'); plt.ylabel('P1_dx (mm)'); plt.title('RSM contour'); plt.show()
```
